Create Custon y and x axes

```python
#####################################
#------ Y-Axis ------#
#####################################
# Each entry: (display label, value in seconds used for data mapping)
y_ticks = [
    ('0.1 sec',  0.1),
    ('1 sec',    1),
    ('1 min',    60),
    ('1 hr',     3600),
    ('1 day',    86400),
    ('1 week',   7*86400),
    ('1 mon',    30*86400),
    ('1 year',   365*86400),
    ('10 yrs',   10*365*86400),
    ('100 yrs',  100*365*86400),
]

# Assign equal linear positions 0, 1, 2, ...
y_labels   = [t[0] for t in y_ticks]
y_seconds  = np.array([t[1] for t in y_ticks])
y_positions = np.arange(len(y_ticks), dtype=float)  # 0, 1, 2, ...

def seconds_to_ypos(s):
    """Interpolate seconds value to linear y-axis position (log-interp between ticks)."""
    log_s   = np.log10(np.clip(s, y_seconds[0], y_seconds[-1]))
    log_sec = np.log10(y_seconds)
    return np.interp(log_s, log_sec, y_positions)

#####################################
#------ X-Axis ------#
#####################################
x_ticks = [
    ('1 mm',       1e-3),
    ('1 cm',       1e-2),
    ('10 cm',      1e-1),
    ('1 m',        1),
    ('10 m',       10),
    ('100 m',      100),
    ('1 km',       1e3),
    ('10 km',      1e4),
    ('100 km',     1e5),
    (r'$10^3$ km', 1e6),
    (r'$10^4$ km', 1e7),
]

# Assign equal linear positions 0, 1, 2, ...
x_labels    = [t[0] for t in x_ticks]
x_metres    = np.array([t[1] for t in x_ticks])
x_positions = np.arange(len(x_ticks), dtype=float)  # 0, 1, 2, ...

def metres_to_xpos(m):
    """Interpolate metres value to linear x-axis position (log-interp between ticks)."""
    log_m   = np.log10(np.clip(m, x_metres[0], x_metres[-1]))
    log_met = np.log10(x_metres)
    return np.interp(log_m, log_met, x_positions)

# Helper functions to convert from physical units to seconds/metres for data input
def to_seconds(value, unit):
    f = {'sec': 1, 'min': 60, 'hr': 3600, 'day': 86400,
         'week': 7*86400, 'month': 30*86400, 'year': 365*86400}
    return value * f[unit]

def to_metres(value, unit):
    f = {'mm': 1e-3, 'cm': 1e-2, 'm': 1, 'km': 1e3}
    return value * f[unit]
```

Plot Stommel Diagram

```python
# Set plotting parameters
fontsize = 16
ROOT = '/Users/lukecolosi/Desktop/projects/graduate_research/Gille_lab/'
PATH_figs =  ROOT + 'AirSeaCoupling/figs/paper_figures/'

# Create figure and axes
fig, ax = plt.subplots(figsize=(11, 8))

# Set figure attributes
ax.set_xlim(-0.5, len(x_positions) - 0.5)
ax.set_ylim(-0.5, len(y_positions) - 0.5)

#--- Major ticks and labels ---
ax.set_xticks(x_positions)
ax.set_xticklabels(x_labels, fontsize=fontsize)
ax.set_yticks(y_positions)
ax.set_yticklabels(y_labels, fontsize=fontsize)

#--- Minor gridlines halfway between each tick --- 
minor_x = np.arange(-0.5, len(x_positions) - 0.5, 1.0)  # 0.5 between each major
minor_y = np.arange(-0.5, len(y_positions) - 0.5, 1.0)

ax.set_xticks(minor_x, minor=True)
ax.set_yticks(minor_y, minor=True)

# Set major and minor gridlines
ax.grid(True, which='minor', color='lightgray', linewidth=0.5, linestyle='--', zorder=0)
ax.tick_params(which='minor', length=0) 
ax.grid(True, which='major', color='lightgray', linewidth=0.5, linestyle='--', zorder=0)
ax.set_axisbelow(True)

ax.set_xlabel('Spatial Scale', fontsize=fontsize+2, labelpad=8)
ax.set_ylabel('Time Scale',    fontsize=fontsize+2, labelpad=8)

#####################################
#------ Shaded region: Model resolution ------#
#####################################
res_x_min = metres_to_xpos(to_metres(1, 'km'))
res_x_max = metres_to_xpos(to_metres(800, 'km'))
res_y_min = seconds_to_ypos(to_seconds(1, 'hr'))
res_y_max = seconds_to_ypos(to_seconds(2, 'year'))

# ax.fill_betweenx(
#     [res_y_min, res_y_max],
#     res_x_min, res_x_max,
#     color='lightblue', alpha=0.4, zorder=1, label ='Model Resolved Scales'
# )

#####################################
#------ Draw Ellipses ------#
#####################################
for name, x_min, x_max, t_min, t_max in processes:
    # Convert physical values → axis positions
    px_min = metres_to_xpos(x_min)
    px_max = metres_to_xpos(x_max)
    py_min = seconds_to_ypos(t_min)
    py_max = seconds_to_ypos(t_max)

    cx = (px_min + px_max) / 2
    cy = (py_min + py_max) / 2
    rx = (px_max - px_min) / 2
    ry = (py_max - py_min) / 2

    theta        = np.linspace(0, 2*np.pi, 300)
    ex0, ey0     = rx * np.cos(theta), ry * np.sin(theta)

    angle_rad    = np.radians(rotations[name])
    cos_a, sin_a = np.cos(angle_rad), np.sin(angle_rad)
    ex = cx + cos_a * ex0 - sin_a * ey0
    ey = cy + sin_a * ex0 + cos_a * ey0

    col = colors[name]
    ax.fill(ex, ey, color=col, alpha=0.1, zorder=2)
    ax.plot(ex, ey, color=col, linewidth=1.2, zorder=3)

    ax.text(cx, cy, latexlabel(name),
            ha='center', va='center',
            fontsize=fontsize-1, color='black',
            linespacing=1.3, zorder=4)

#####################################
#------ Spines ------#
#####################################
for spine in ax.spines.values():
    spine.set_linewidth(1.2)

# Add legend for shaded region
#ax.legend(loc='upper left', fontsize=fontsize-2)

# Adjust layout and show figure
plt.tight_layout()
plt.show()

# Save with high quality
figName = 'fig12.png'

fig.savefig(
    PATH_figs + figName,
    dpi=300,
    facecolor='white',
    bbox_inches='tight',
    pad_inches=0.1,
    transparent=False
)
```
